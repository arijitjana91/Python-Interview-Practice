# Q1. Count occurrence of each character in a string.

## Method 1 — Using Dictionary

```python

def char_count(string):
    count_dict = {}
    string = string.replace(" ","")
    string = string.lower()
    for char in string:
        if char in count_dict:
            count_dict[char]+= 1
        else:
            count_dict[char] = 1
    return count_dict
        


string = "dataengineering"
print(char_count(string))
```

## Method 2 — Using .get() method

```python

def char_count(string):
    char_count = {}
    string = string.replace(" ","")
    string = string.lower()
    for char in string:
        char_count[char] = char_count.get(char, 0) + 1
    return char_count


string = "Data Engineering"
print(char_count(string))
```

## Method 3 — Using collections.Counter

```python
from collections import Counter

# Counter : it is a dictionary subclass for counting hashable objects. It is a collection where elements are stored as dictionary keys and their counts are stored as dictionary values. Counts are allowed to be any integer value including zero or negative counts. The Counter class is similar to bags or multisets in other languages.

string = "dataengineering"
result = Counter(string)
print(result)
```

## Method 4 — Using PySpark RDD (Very Important for Data Engineering)

> First we convert each element into a key-value pair using map transformation.
Then we use reduceByKey to group values by key and aggregate them using addition.
Finally, collect action brings the result from distributed executors to the driver program.

```python
from pyspark import SparkContext

sc = SparkContext.getOrCreate()

string = "dataengineering"

rdd = sc.parallelize(list(string))

print(rdd.collect())
print(rdd.map(lambda x: (x, 1)).collect())

result = (rdd
          .map(lambda x: (x, 1))
          .reduceByKey(lambda a, b: a + b)
          .collect())

print(result)
```

# Q2. Count occurrence of each word in a sentence.

```python

def word_count(sentence):
  count_dict = {}

  for word in sentence.split(" "):
    if word in count_dict:
      count_dict[word]+= 1
    else:
      count_dict[word] = 1
  return count_dict

sentence = "data engineering is fun data engineering is powerful"
print(word_count(sentence))
```

```python
from collections import Counter

result = Counter(sentence.split(" "))

print(result)
```

```python
from pyspark import SparkContext

sc = SparkContext.getOrCreate()

sentence = "data engineering is fun data engineering is powerful"

rdd = sc.parallelize(sentence.split(" "))

print("Rdd Data: ",rdd.collect())

result = (rdd
            .map(lambda x: (x,1))
            .reduceByKey(lambda a,b: a+b))

print(result.collect())
```

# Pandas Solution (Sometimes Asked)
